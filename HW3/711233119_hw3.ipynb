```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import f, probplot

# ECDF
def ecdf(sample):
    x = np.sort(sample)
    y = np.arange(1, len(x) + 1) / len(x)
    return x, y

# 
d1, d2 = 5, 10

# 
sample_sizes = [30, 60, 200, 1000]

# 
plt.figure(figsize=(12, 8))

for i, n in enumerate(sample_sizes):
    #  F 
    sample = np.random.f(d1, d2, size=n)

    # 
    plt.subplot(2, 4, i + 1)
    plt.hist(sample, bins=20, density=True, alpha=1, color='b', edgecolor='y', linewidth=0.8)
    x_range = np.linspace(0, sample.max(), 100)
    plt.plot(x_range, f.pdf(x_range, d1, d2), 'r', label='Theoretical PDF')
    plt.title(f'Histogram (n={n})')
    plt.legend()

    # ECDF
    plt.subplot(2, 4, i + 5)
    x_ecdf, y_ecdf = ecdf(sample)
    plt.step(x_ecdf, y_ecdf, color='g', linewidth=1.5)
    x_range = np.linspace(0, sample.max(), 100)
    y_range = f.cdf(x_range, d1, d2)
    plt.plot(x_range, y_range, 'r', label='Theoretical CDF')
    plt.title(f'Empirical CDF (n={n})')
    plt.legend()

plt.tight_layout()
plt.savefig('/Users/gp/Desktop/XeLaTex/HW3/images/F-distribution_sample.png', format='png', dpi=300)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Matplotlib
plt.style.use('default')

def plot_sample_distribution(sample_size_list):
    plt.figure(figsize=(12, 6))
    colors = ['c', 'g', 'orange', 'b']
    for i, sample_size in enumerate(sample_size_list):
        sample_data = np.random.normal(0, 1, sample_size)
        plt.hist(sample_data, bins=30, density=True, alpha=0.6, label=f'Sample Size {sample_size}', color=colors[i])

    x = np.linspace(-4, 4, 1000)
    plt.plot(x, stats.norm.pdf(x, 0, 1), 'k', label='Theoretical Distribution', color='r', linewidth=3 )
    #plt.title('Sample Distribution of Normal Distribution with Varying Sample Size')
    plt.legend(fontsize=14)
    plt.tight_layout()
    plt.grid(True)
    plt.savefig('/Users/gp/Desktop/XeLaTex/HW3/images/normal-distribution_sample.png', format='png', dpi=300)
    plt.show()

sample_size_list = [10, 100, 1000, 10000]
plot_sample_distribution(sample_size_list)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

def simulate_exponential(lambda_, size):
    return np.random.exponential(1/lambda_, size)



def plot_simulation(lambda_, size_list):
    fig, axs = plt.subplots(4, len(size_list), figsize=(15, 15))

    for i, size in enumerate(size_list):
        samples = simulate_exponential(lambda_, size)

        # Boxplot
        axs[0, i].boxplot(samples)
        axs[0, i].set_title(f"Boxplot for size={size}")

        # QQplot
        stats.probplot(samples, dist=stats.expon(scale=1/lambda_), plot=axs[1, i], rvalue=True)
        axs[1, i].set_title(f"QQplot for size={size}")

        # ECDF
        x = np.sort(samples)
        y = np.arange(1, size+1) / size
        axs[2, i].step(x, y, where='post', label='Empirical', color='b', linewidth=2)
        
        # Adding theoretical CDF
        t_x = np.linspace(0, np.max(samples), 400)
        t_y = stats.expon.cdf(t_x, scale=1/lambda_)
        axs[2, i].plot(t_x, t_y, color='red', linestyle='--', label='Theoretical')
        
        axs[2, i].set_title(f"ECDF for size={size}")
        axs[2, i].legend()

        # Comparison with theoretical (PDF vs Theoretical PDF)
        axs[3, i].hist(samples, bins=40, density=True, alpha=0.7, label='Simulation', color='b', edgecolor='y', linewidth=0.8)
        t_x = np.linspace(0, np.max(samples), 400)
        t_y = stats.expon.pdf(t_x, scale=1/lambda_)
        axs[3, i].plot(t_x, t_y, color='red', lw=2, label='Theoretical')
        axs[3, i].set_title(f"Comparison for size={size}")
        axs[3, i].legend()

    plt.tight_layout()
    plt.savefig('/Users/gp/Desktop/XeLaTex/HW3/images/rnb_exponential_distribution.png', format='png', dpi=300)
    plt.show()

lambda_ = 2  
size_list = [10, 100, 10000]  # 
plot_simulation(lambda_, size_list)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# ECDF
def ecdf(sample):
    x = np.sort(sample)
    y = np.arange(1, len(x) + 1) / len(x)
    return x, y

# 
n_trials = 20
p_success = 0.3

# 
sample_sizes = [30, 60, 200, 1000]

# 
plt.figure(figsize=(12, 8))

for i, n in enumerate(sample_sizes):
    # 
    sample = np.random.binomial(n_trials, p_success, size=n)

    # 
    plt.subplot(2, 4, i + 1)
    plt.hist(sample, bins=np.arange(0, n_trials+1) - 0.5, density=True, alpha=1, color='b', edgecolor='y', linewidth=0.8)
    x_range = np.arange(0, n_trials + 1)
    plt.plot(x_range, binom.pmf(x_range, n_trials, p_success), 'r', marker='o', linestyle='-', label='Theoretical PMF')
    plt.title(f'Histogram (n={n})')
    plt.legend()

    # ECDF
    plt.subplot(2, 4, i + 5)
    x_ecdf, y_ecdf = ecdf(sample)
    plt.step(x_ecdf, y_ecdf, color='g', linewidth=1.5)
    x_range = np.arange(0, n_trials + 1)
    y_range = binom.cdf(x_range, n_trials, p_success)
    plt.plot(x_range, y_range, 'r', marker='o', linestyle='-', label='Theoretical CDF')
    plt.title(f'Empirical CDF (n={n})')
    plt.legend()

plt.tight_layout()
plt.savefig('/Users/gp/Desktop/XeLaTex/HW3/images/binomial-distribution_sample.png', format='png', dpi=300)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
from scipy.stats import skew

# 
sample_sizes = [10, 20, 30, 50, 100, 300, 500]
num_simulations = 10000

# 
simulated_G1 = {}

# 
for n in sample_sizes:
    G1_samples = []
    
    for _ in range(num_simulations):
        # 
        samples = np.random.normal(0, 1, n)
        
        # 
        skew_estimate = skew(samples)
        
        #  G1
        G1 = np.sqrt(n / 6) * skew_estimate
        G1_samples.append(G1)
    
    simulated_G1[n] = G1_samples

#  n = 10  n = 500  ECDF 
plt.figure(figsize=(12, 6))

for n in [10, 500]:
    plt.subplot(1, 2, n//490+1)
    
    # 
    plt.hist(simulated_G1[n], bins=50, density=True, alpha=1, label=f'n={n}',color='c')
    
    #  PDF
    x_range = np.linspace(-5, 5, 1000)
    pdf = norm.pdf(x_range, loc=0, scale=1)
    plt.plot(x_range, pdf, 'r', label='Standard Normal PDF', linewidth=2)
    plt.legend()
    #  CDF
    cdf_values = norm.cdf(x_range, loc=0, scale=1)
    
    #  CDF
    plt.twinx()
    plt.plot(x_range, cdf_values, 'b', label='Standard Normal CDF', linewidth=2)
    
    #plt.title(f'Simulation vs. Standard Normal (n={n})')
    #plt.xlabel("G1")
    #plt.ylabel("Density / CDF")
    plt.legend()
    
plt.tight_layout()
plt.savefig('/Users/gp/Desktop/XeLaTex/HW3/images/sampling-distribution_G1.png', format='png', dpi=300)
plt.show()
```
