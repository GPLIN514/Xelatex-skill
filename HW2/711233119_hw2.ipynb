```python
import numpy as np
import matplotlib.pyplot as plt

# 定義函數
def sigmoid(x, alpha):
    return (np.exp(alpha * x)) / (np.exp(alpha * x)+1)

# 產生X值範圍
x = np.linspace(-6, 6, 400)

# 使用 alpha=1 繪製 sigmoid 函數
alpha = 1
y = sigmoid(x, alpha)
plt.plot(x, y, label=f'$\\alpha = {alpha}$',color='b')

plt.axhline(1, color='red', linestyle='--', label='Asymptote at y=1')

ax = plt.gca()
ax.set_facecolor('#D5FAF9')
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import argrelextrema

x = np.linspace(-9 * np.pi, 3 * np.pi, 200)
y = np.exp(-x / 10) * np.sin(x)

fig, ax = plt.subplots(1)
ax.plot(x, y, color="b", label='f(x)')
ax.set_xticks(np.array([-9, -8, -7, -6, -5, -4, -3, -2, -1, 0, 1, 2, 3]) * np.pi)
ax.set_xticklabels(
    ["-9$\pi$", "-8$\pi$", "-7$\pi$", "-6$\pi$", "-5$\pi$", "-4$\pi$", "-3$\pi$", "-2$\pi$", "$\pi$", "0", "$\pi$", "2$\pi$", "3$\pi$"],
    fontsize=10,
    color="b",
)

ax.set_facecolor('#D5FAF9')
ax.set_ylim([-17, 15])
ax.set_xlabel('x')
ax.set_ylabel('f(x)')
ax.grid(True)

# Find and plot local maxima
local_maxima_indices = argrelextrema(y, np.greater)
local_maxima_x = x[local_maxima_indices]
local_maxima_y = y[local_maxima_indices]
ax.scatter(local_maxima_x, local_maxima_y, color='r', label='Local Maxima', s=30, zorder=2)

# Find and plot local minima
local_minima_indices = argrelextrema(y, np.less)
local_minima_x = x[local_minima_indices]
local_minima_y = y[local_minima_indices]
ax.scatter(local_minima_x, local_minima_y, color='g', label='Local Minima', s=30, zorder=2)

plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import argrelextrema

x = np.linspace(-9 * np.pi, 3 * np.pi, 200)
y = np.exp(-x / 10) * np.sin(x)

fig, ax = plt.subplots(1)
ax.plot(x, y, color="b", label='f(x)')
ax.set_xticks(np.array([-9, -8, -7, -6, -5, -4, -3, -2, -1, 0, 1, 2, 3]) * np.pi)
ax.set_xticklabels(
    ["-9$\pi$", "-8$\pi$", "-7$\pi$", "-6$\pi$", "-5$\pi$", "-4$\pi$", "-3$\pi$", "-2$\pi$", "$\pi$", "0", "$\pi$", "2$\pi$", "3$\pi$"],
    fontsize=10,
    color="b",
)

ax.set_facecolor('#D5FAF9')
ax.set_ylim([-17, 15])
ax.set_xlabel('x')
ax.set_ylabel('f(x)')
ax.grid(True)

# Find and plot local maxima
local_maxima_indices = argrelextrema(y, np.greater)
local_maxima_x = x[local_maxima_indices]
local_maxima_y = y[local_maxima_indices]
ax.scatter(local_maxima_x, local_maxima_y, color='r', label='Local Maxima', s=30, zorder=2)

# Find and plot local minima
local_minima_indices = argrelextrema(y, np.less)
local_minima_x = x[local_minima_indices]
local_minima_y = y[local_minima_indices]
ax.scatter(local_minima_x, local_minima_y, color='g', label='Local Minima', s=30, zorder=2)

# Fit and plot curves for local maxima and minima
maxima_curve = np.polyfit(local_maxima_x, local_maxima_y, 5)  # Adjust the degree as needed
minima_curve = np.polyfit(local_minima_x, local_minima_y, 5)  # Adjust the degree as needed
x_curve = np.linspace(-9 * np.pi, 3 * np.pi, 1000)
y_maxima_curve = np.polyval(maxima_curve, x_curve)
y_minima_curve = np.polyval(minima_curve, x_curve)

ax.plot(x_curve, y_maxima_curve, color='r', label='Maxima Curve', linestyle='-', linewidth=2)
ax.plot(x_curve, y_minima_curve, color='g', label='Minima Curve', linestyle='-', linewidth=2)

plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 定義高斯分佈的函數
def gaussian(x, mu, sigma):
    return 1 / (sigma * np.sqrt(2 * np.pi)) * np.exp(-0.5 * ((x - mu) / sigma) ** 2)

# 產生 x 範圍
x = np.linspace(-4, 10, 400)

# 計算高斯分佈的值，使用 mu=3 和 sigma=1
mu = 3
sigma = 1
y = gaussian(x, mu, sigma)

# 繪製高斯分佈曲線
plt.plot(x, y, label=f'$N(\mu={mu}, \sigma={sigma})$', color="blue")

# 使用 fill_betweenx 函數繪製不同顏色的區域
plt.fill_between(x, 0, y, where=(x >= mu - sigma) & (x <= mu + sigma), color='#00FA9A', alpha=0.4)
plt.fill_between(x, 0, y, where=(x >= mu - 2 * sigma) & (x <= mu - sigma), color='yellow', alpha=0.4)
plt.fill_between(x, 0, y, where=(x >= mu + sigma) & (x <= mu + 2 * sigma), color='yellow', alpha=0.4)
plt.fill_between(x, 0, y, where=(x >= mu - 3 * sigma) & (x <= mu - 2 * sigma), color='red', alpha=0.4)
plt.fill_between(x, 0, y, where=(x >= mu + 2 * sigma) & (x <= mu + 3 * sigma), color='red', alpha=0.4)


plt.axvline(x=mu - sigma, color='black', linestyle='-', linewidth=1, ymin=0.045,ymax=0.6)
plt.axvline(x=mu + sigma, color='black', linestyle='-', linewidth=1, ymin=0.045,ymax=0.6)
plt.axvline(x=mu - 2 * sigma, color='black', linestyle='-', linewidth=1, ymin=0.045,ymax=0.165)
plt.axvline(x=mu + 2 * sigma, color='black', linestyle='-', linewidth=1, ymin=0.045,ymax=0.165)
plt.axvline(x=mu - 3 * sigma, color='black', linestyle='-', linewidth=1, ymin=0.045,ymax=0.055)
plt.axvline(x=mu + 3 * sigma, color='black', linestyle='-', linewidth=1, ymin=0.045, ymax=0.055)


plt.annotate('68.2%', (mu, 0.15), textcoords="offset points", xytext=(0, 10), ha='center', fontsize=12)


ax=plt.gca()
#  設置了底部坐標軸的位置，是在x軸的數值0位置。這確保底部坐標軸穿過y=0這一點。
ax.spines['bottom'].set_position(('data',0))
#  設置了頂部坐標軸的可見性，將頂部的坐標軸設置為不可見，也就是隱藏它。
ax.spines['top'].set_visible(False) # take off top line
#  設置了右側坐標軸的可見性，將右側的坐標軸設置為不可見，也就是隱藏它。
ax.spines['right'].set_visible(False)

ax.annotate('13.6%', \
    xy=(1.6, 0.05), xytext=(-0.1, 0.15), \
        arrowprops=dict(arrowstyle='-|>', \
            connectionstyle='arc3, rad=0.3'), fontsize =12)
ax.annotate('13.6%', \
    xy=(4.5, 0.05), xytext=(4.5, 0.15), \
        arrowprops=dict(arrowstyle='-|>', \
            connectionstyle='arc3, rad=-0.3'), fontsize =12)
ax.annotate('2.1%', \
    xy=(0.9, 0.01), xytext=(-3, 0.05), \
        arrowprops=dict(arrowstyle='-|>', \
            connectionstyle='arc3, rad=-0.3'), fontsize =12)
ax.annotate('2.1%', \
    xy=(5.1, 0.01), xytext=(8, 0.05), \
        arrowprops=dict(arrowstyle='-|>', \
            connectionstyle='arc3, rad=0.3'), fontsize =12)

#  設置x軸的刻度位置
ax.set_xticks(np.array([ -4,-2,0, 1, 2, 3, 4, 5, 6,8,10]))
#  設置x軸的刻度標籤，同時設置字體大小為10，標籤顏色為藍色。
ax.set_xticklabels(
    [-4,-2,0,1,2,3,4,5,6,8,10],
    fontsize=10,
    color="b",)

plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 定義隱函數
def implicit_circle(x, y):
    return x**2 + y**2 - 1

# 產生 x 和 y 座標的數據點
x = np.linspace(-1.5, 1.5, 400)
y = np.linspace(-1.5, 1.5, 400)
X, Y = np.meshgrid(x, y)
Z = implicit_circle(X, Y)

# 繪製圓形方程式的圖形
plt.figure(figsize=(6, 6))
plt.contour(X, Y, Z, levels=[0], colors='b')  # levels=[0] 表示畫出等於 0 的等高線
plt.gca().set_aspect('equal', adjustable='box')  # 讓 x 和 y 軸的比例相同，確保圓形正確顯示

# 設定座標軸的範圍，讓它們置中
plt.gca().set_aspect('equal', adjustable='box')  # 讓 x 和 y 軸的比例相同，確保圓形正確顯示
plt.gca().spines['left'].set_position('center')
plt.gca().spines['bottom'].set_position('center')
plt.gca().spines['right'].set_color('none')
plt.gca().spines['top'].set_color('none')
ax = plt.gca()
ax.set_facecolor('#D5FAF9')

plt.xlim(-1.5, 1.5)
plt.ylim(-1.5, 1.5)
plt.grid(True)
plt.show()
```

\begin{align*}
\gamma_n &=\sum_{i=1}^n\int_{i}^{i+1} \int_{\frac{1}{x}}^{\frac{1}{i}}dydx \\
         &=\sum_{i=1}^n\int_{i}^{i+1} \frac{1}{i}-\frac{1}{x}dydx \\
         &=\sum_{i=1}^n\left(\left. \frac{x}{i}-ln(x)\right|_{i}^{i+1}\right)\\
         &=\sum_{i=1}^n\left( \frac{i+1}{i}-ln(i+1)-\frac{i}{i}+ln(i)\right) \\
         &=\sum_{i=1}^n\left( \frac{1}{i}-ln\left(\frac{i+1}{i}\right)\right)  \\
         &=\sum_{i=1}^n\left(\frac{1}{i}\right) -\sum_{i=1}^nln\left(\frac{i+1}{i}\right) \\
         &=S_n-\left[ln\left(\frac{2}{1}\right)+ln\left(\frac{3}{2}\right)+ln\left(\frac{4}{3}\right)+\cdots+ln\left(\frac{n+1}{n}\right)\right]\\
         &=S_n-ln\left(\frac{2}{1}\cdot \frac{3}{2}\cdot \frac{4}{3}\cdot \cdots \cdot \frac{n+1}{n}\right)\\
         &=S_n-ln\left(n+1\right)\\
\end{align*}

```python
import numpy as np
import matplotlib.pyplot as plt

t = np.arange(0.1, 10, 0.1)
h = 1/t

fig, ax = plt.subplots()

ax.plot(t, h, color='b')
ax.set_ylim(0, 2)

for i in range(0, 10):
    x = np.arange(1 + i, 2 + i, 0.1)
    y = 1/x
    g = [1/(i + 1)] * 10
    ax.fill_between(x, y, g, color='#FF8C00', alpha=0.5)

ax.text(1.0,1.5,'$f(x)=\\frac{1}{x}$',fontsize='12')
ax.text(5,0.75,'$\gamma_n$ = orange area',fontsize='12')
ax.set_facecolor('#D5FAF9')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

colors = ["r", "b", "g"]
x = np.linspace(0, 100)
fig, ax = plt.subplots()

y1 = np.ones(x.shape)  # Create an array of 1s with the same shape as x
ax.plot(x, y1, label="$1$", color=colors[0])

n = np.arange(1, len(x) + 1)  # Create an array of n values from 1 to the length of x

y2 = np.cumsum(1 / n) - np.log(n + 1)
ax.plot(x, y2, label="$\gamma _n$", color=colors[1])

y3 = 0.5 * (1 - 1 / (n + 1))
ax.plot(x, y3, label="$\\frac{1}{2}(1-\\frac{1}{n+1})$", color=colors[2])

ax.set_facecolor('#D5FAF9')
ax.legend()
ax.grid(True)
plt.xlabel("n")
ax.set_ylim([0.2, 1.1])
ax.set_xticks(np.array([0, 20, 40, 60, 80, 100]))  # Adjust x-ticks
ax.set_xticklabels(
    ["0", "20", "40", "60", "80", "100"],
    fontsize=10,
    color="b"
)
plt.show()

```

```python
import matplotlib.pyplot as plt
import numpy as np

n = 8
side_length = 2  
center = (4, 0)  
rotation_angle = np.deg2rad(360/n) 

square_x = [center[0] - side_length / 2, center[0] + side_length / 2, center[0] + side_length / 2, center[0] - side_length / 2, center[0] - side_length / 2,center[0]-side_length/2]
square_y = [center[1] - side_length / 2, center[1] - side_length / 2, center[1] + side_length / 2, center[1] + side_length / 2, center[1] - side_length / 2,center[1]]
square = np.array([square_x, square_y])
colors = ['blue', 'red', 'green', 'orange', 'purple', 'pink', 'brown', 'gray']
fig, ax = plt.subplots(figsize=(4, 4))
ax.set_xlim(-6, 6)
ax.set_ylim(-6, 6)
ax.set_aspect(1)
for _ in range(n):
    rotation_matrix = np.array([[np.cos(rotation_angle), -np.sin(rotation_angle)], [np.sin(rotation_angle), np.cos(rotation_angle)]])
    square = np.dot(rotation_matrix, square)
    colort = colors[_ % len(colors)]
    ax.plot(square[0], square[1],lw=1, color=colort)
    ax.plot([square[0][5], 0], [square[1][5], 0],lw=1,   color=colort)
plt.savefig(fname='ferris' , dpi=300,bbox_inches='tight')
plt.show()
```
